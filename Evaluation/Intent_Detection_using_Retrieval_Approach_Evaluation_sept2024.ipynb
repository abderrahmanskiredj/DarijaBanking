```python
from collections import Counter

def compute_metrics(ground_truth, predictions):
    unique_labels = set(ground_truth + predictions)
    true_positives = Counter()
    false_positives = Counter()
    false_negatives = Counter()

    for true_label, predicted_label in zip(ground_truth, predictions):
        if true_label == predicted_label:
            true_positives[true_label] += 1
        else:
            false_positives[predicted_label] += 1
            false_negatives[true_label] += 1

    precision = {}
    recall = {}
    f1_score = {}

    for label in unique_labels:
        precision[label] = true_positives[label] / (true_positives[label] + false_positives[label]) if (true_positives[label] + false_positives[label]) else 0
        recall[label] = true_positives[label] / (true_positives[label] + false_negatives[label]) if (true_positives[label] + false_negatives[label]) else 0
        f1_score[label] = 2 * (precision[label] * recall[label]) / (precision[label] + recall[label]) if (precision[label] + recall[label]) else 0

    macro_precision = sum(precision.values()) / len(unique_labels)
    macro_recall = sum(recall.values()) / len(unique_labels)
    macro_f1_score = sum(f1_score.values()) / len(unique_labels)
    accuracy = sum(true_positives.values()) / len(ground_truth)

    return accuracy, macro_precision, macro_recall, macro_f1_score

# Example usage:
ground_truth = ['a', 'b', 'c', 'a', 'b', 'c']
predictions = ['a', 'b', 'c', 'c', 'b', 'a']
accuracy, precision, recall, f1_score = compute_metrics(ground_truth, predictions)
print(f"Accuracy: {accuracy}\nPrecision: {precision}\nRecall: {recall}\nF1 Score: {f1_score}")
```

# Using Arabic LMs as retrievers

```python
from tqdm import tqdm
import numpy as np

def encode(texts, model, tokenizer):
    texts_batched = [texts[i:i+20] for i in range(0, len(texts), 20)]
    results_batched = []
    for batch in tqdm(texts_batched):
        with torch.no_grad():
            inputs = tokenizer(batch, return_tensors="pt", padding=True)
            outputs = model(**inputs)
        results_batched.append(outputs.last_hidden_state)
    embeddings = []
    for batch in results_batched:
        for seq in batch:
            embeddings.append(seq[0].tolist())
    return embeddings
```

```python
ground_truths = df['label'].tolist()
```

```python
from transformers import BertModel, BertTokenizer
import torch
from IPython.display import display
import faiss


model_names = [
 'SI2M-Lab/DarijaBERT',
                   'alger-ia/dziribert',
                  'aubmindlab/bert-base-arabertv02-twitter',
                   "qarib/bert-base-qarib","CAMeL-Lab/bert-base-arabic-camelbert-da", 'CAMeL-Lab/bert-base-arabic-camelbert-mix','UBC-NLP/MARBERT', 'UBC-NLP/ARBERT',
                   'UBC-NLP/MARBERTv2', 'UBC-NLP/ARBERTv2']

dres = pd.DataFrame(columns=['model name', "accuracy msa", "precision msa", "recall msa", "f1 msa", "accuracy darija", "precision darija", "recall darija", "f1 darija"])
for model_name in model_names:
    print(model_name)
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertModel.from_pretrained(model_name, add_pooling_layer=False)
    embeddings = encode(df['text'].tolist(), model, tokenizer)
    embeddings_test_dar_normed = encode(dtest_darija['text'].tolist(),model, tokenizer)
    embeddings_test_ar_normed = encode(dtest_arabe['text'].tolist(),model, tokenizer)
    embeddings = np.array(embeddings)
    embeddings_test_ar_normed = np.array(embeddings_test_ar_normed)
    embeddings_test_dar_normed = np.array(embeddings_test_dar_normed)
    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)

    query = embeddings_test_ar_normed
    #y = y_test_darija
    k = 1 # number of nearest neighbors to return
    D, I = index.search(query, k)
    predicted_intents_for_ar = [ground_truths[x[0]] for x in I.tolist()]

    index = faiss.IndexFlatL2(embeddings.shape[1])
    index.add(embeddings)

    query = embeddings_test_dar_normed
    #y = y_test_darija
    k = 1 # number of nearest neighbors to return
    D, I = index.search(query, k)
    predicted_intents_for_dar = [ground_truths[x[0]] for x in I.tolist()]
    ground_truths_test_ar = dtest_arabe['label'].tolist()
    ground_truths_test_dar = dtest_darija['label'].tolist()
    acc_msa, prec_msa, rec_msa, f1_msa = compute_metrics(ground_truths_test_ar, predicted_intents_for_ar)
    acc_dar, prec_dar, rec_dar, f1_dar = compute_metrics(ground_truths_test_dar, predicted_intents_for_dar)
    new_row = {'model name': model_name, "accuracy msa": acc_msa, "precision msa":prec_msa,  "recall msa":rec_msa,
               "f1 msa": f1_msa, "accuracy darija": acc_dar, "precision darija": prec_dar, "recall darija": rec_dar, "f1 darija": f1_dar}

    # create a new DataFrame with the new row data
    new_df = pd.DataFrame(new_row, index=[0])

    # concatenate the two DataFrames using the pandas.concat function
    dres = pd.concat([dres, new_df], ignore_index=True)
    display(dres)
```
